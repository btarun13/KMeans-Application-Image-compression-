```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [15,10] #(6,4)

im = cv2.imread('/home/tarun/Pictures/landscape.jpg')
im.shape
#(length,width,color or channel)
```

```python
(h,w,c) = im.shape
plt.imshow(im)
plt.show()
```

```python
image = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

plt.imshow(image)
plt.show()

plt.imshow(np.hstack([im,image]))
plt.show()
```

```python
image = image.reshape(-1,3)
image.shape
```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.cluster import KMeans
```

```python
Num_cluster = 16
clt = MiniBatchKMeans(n_clusters=Num_cluster)
labels = clt.fit_predict(image) 
```

```python
quant = clt.cltuster_centers_.astype('uint8')[labels]
quant.shape
```

```python
quant = quant.reshape(h,w,3)
plt.imshow(quant)
```
